```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv("sales_data_sample.csv", encoding='latin1')
```

```python
df.columns
```

```python
# Scatter plot
plt.scatter(df['QUANTITYORDERED'], df['SALES'])

# Adding labels and title
plt.xlabel('Quantity Ordered')
plt.ylabel('Sales')
plt.title('Scatter Plot: Quantity Ordered vs Sales')

# Show the plot
plt.show()
```

```python
# Hexbin plot
plt.hexbin(df['QUANTITYORDERED'], df['SALES'], gridsize=20, cmap='Blues')

# Adding labels and title
plt.xlabel('Quantity Ordered')
plt.ylabel('Sales')
plt.title('Hexbin Plot: Quantity Ordered vs Sales')

# Adding a colorbar
plt.colorbar(label='Count')

# Show the plot
plt.show()
```

```python
# Bar plot
plt.bar(df['QUANTITYORDERED'], df['SALES'], color='skyblue')

# Adding labels and title
plt.xlabel('Quantity Ordered')
plt.ylabel('Sales')
plt.title('Bar Plot: Quantity Ordered vs Sales')

# Show the plot
plt.show()
```

```python
# Violin plot
sns.violinplot(x='QUANTITYORDERED', y='SALES', data=df, palette='pastel')

# Adding labels and title
plt.xlabel('Quantity Ordered')
plt.ylabel('Sales')
plt.title('Violin Plot: Quantity Ordered vs Sales')

# Show the plot
plt.show()
```

```python
# Selecting relevant columns
relevant_cols = df[['QUANTITYORDERED', 'SALES']]

# Pair plot
sns.pairplot(relevant_cols, diag_kind='kde')

# Show the plot
plt.show()
```

```python
# Radial plot
theta = np.linspace(0, 2*np.pi, len(df))
radius = df['SALES']
plt.polar(theta, radius)

# Adding labels and title
plt.title('Radial Plot: Quantity Ordered vs Sales')

# Show the plot
plt.show()
```

```python
# 3D Scatter plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df['QUANTITYORDERED'], df['SALES'], c='skyblue', marker='o')

# Adding labels and title
ax.set_xlabel('Quantity Ordered')
ax.set_ylabel('Sales')
ax.set_zlabel('Z-axis')
ax.set_title('3D Scatter Plot: Quantity Ordered vs Sales')

# Show the plot
plt.show()
```

```python
# Create a density plot
plt.figure(figsize=(10, 6))
sns.kdeplot(df['QUANTITYORDERED'], label='Quantity Ordered', fill=True)
sns.kdeplot(df['SALES'], label='Sales', fill=True)

# Adding labels and title
plt.xlabel('Values')
plt.ylabel('Density')
plt.title('Density Plot: Quantity Ordered and Sales')

# Adding legend
plt.legend()

# Show the plot
plt.show()
```
